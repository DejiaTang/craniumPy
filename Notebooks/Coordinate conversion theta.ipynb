```python
def read_csvs(filedir):
    files = os.listdir(filedir)
    dfs = []
    for file in files:
        if 'Probabilities' in file and 'csv' in file:
            df = pd.read_csv(os.path.join(filedir,file))
            dfs.append(df)
    return(dfs)
```

```python
def find_range(dfs,key):
    Lmin,Lmax = [],[]
    for df in dfs:
        Lmin.append(np.min(df[key]))
        Lmax.append(np.max(df[key]))
        
    minv = np.min(Lmin)
    maxv = np.max(Lmax)
    return(minv,maxv)
```

```python
def create_avg_hm(dfs,xkey,ykey,xnum,ynum):
    xmin,xmax = find_range(dfs,xkey)
    xbins = np.linspace(xmin,xmax,num=xnum)
    ymin,ymax = find_range(dfs,ykey)
    ybins = np.linspace(ymin,ymax,num=ynum)
    print(np.around(xbins,decimals=1))
    print(np.around(ybins,decimals=1))
    
    Lhm = []
    for i,df in enumerate(dfs):
        hm = np.histogram2d(df[xkey],df[ykey],bins=[xbins,ybins])
        if i == 0:
            hms = np.zeros((len(dfs),hm[0].shape[0],hm[0].shape[1]))
            hms[i] = hm[0]
        else:
            hms[i] = hm[0]
        Lhm.append(hm)
    return(hms,hm[1],hm[2])
```
